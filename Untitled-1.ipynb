```python
import pandas as pd
orders = pd.read_csv('olist_orders_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
print(orders.head())
```

```python
print(payments.head())
```

```python
print(orders.info())
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
merged_data = pd.merge(orders, payments, on='order_id', how='inner')
columns_to_show = ['order_id', 'order_purchase_timestamp', 'payment_value']
print(merged_data[columns_to_show].head())
```

```python
merged_data.set_index('order_purchase_timestamp', inplace=True)
daily_sales = merged_data['payment_value'].resample('D').sum().reset_index()
daily_sales.columns = ['Date', 'Total_Sales']
print(daily_sales.head())
```

```python
import plotly.express as px
fig = px.line(daily_sales, 
              x='Date', 
              y='Total_Sales', 
              title='Olist Store: Daily Total Sales',
              labels={'Date': 'Order Date', 'Total_Sales': 'Total Sales (BRL)'})
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all", label="All")
        ])
    )
)
fig.show()
```

```python
prophet_df = daily_sales.rename(columns={'Date': 'ds', 'Total_Sales': 'y'})
print(prophet_df.head())
```

```python
from prophet import Prophet
model = Prophet()
model.fit(prophet_df)
future_dates = model.make_future_dataframe(periods=90)
forecast = model.predict(future_dates)
columns_to_show = ['ds', 'yhat', 'yhat_lower', 'yhat_upper']
print("Forecast for the next 90 Days ")
print(forecast[columns_to_show].tail())
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
fig_forecast = plot_plotly(model, forecast)
fig_forecast.update_layout(title="Olist Store: 90-Day Sales Forecast")
fig_forecast.show()
fig_components = plot_components_plotly(model, forecast)
fig_components.update_layout(title="Olist Store: Sales Trends and Seasonality Components")
fig_components.show()
```

```python
from prophet.diagnostics import cross_validation, performance_metrics
df_cv = cross_validation(model, initial='365 days', period='30 days', horizon='90 days')
df_p = performance_metrics(df_cv)
columns_to_show = ['horizon', 'mae', 'rmse']
print("--- Model Evaluation Metrics ---")
print(df_p[columns_to_show].tail())
```

```python
order_items = pd.read_csv('olist_order_items_dataset.csv')
freight_data = pd.merge(order_items[['order_id', 'freight_value']], 
                        orders[['order_id', 'order_purchase_timestamp']], 
                        on='order_id')
freight_data.set_index('order_purchase_timestamp', inplace=True)
daily_freight = freight_data['freight_value'].resample('D').mean().fillna(0).reset_index()
daily_freight.columns = ['ds', 'freight_value']
multivariate_df = pd.merge(prophet_df, daily_freight, on='ds', how='inner')
print(" Multivariate Dataset Preview ")
print(multivariate_df.head())
```

```python
# Check the correlation between Sales (y) and Freight Value
correlation = multivariate_df['y'].corr(multivariate_df['freight_value'])
print(f"Correlation between Sales and Freight: {correlation:.2f}")
```

```python
# 1. Aggregate price and unique sellers per order
# We use 'nunique' for seller_id to count how many distinct sellers were active
daily_features = order_items.groupby('order_id').agg({
    'price': 'sum',
    'seller_id': 'nunique'
}).reset_index()

# 2. Merge with orders to get the timestamps
daily_features = pd.merge(daily_features, orders[['order_id', 'order_purchase_timestamp']], on='order_id')

# 3. Resample to Daily frequency
# Sum for price (Total Revenue) and Mean for seller diversity per day
df_advanced = daily_features.set_index('order_purchase_timestamp').resample('D').agg({
    'price': 'mean',
    'seller_id': 'sum'
}).fillna(0).reset_index()

df_advanced.columns = ['ds', 'avg_price', 'active_sellers']

# 4. Merge with our main multivariate_df
multivariate_df = pd.merge(multivariate_df, df_advanced, on='ds', how='inner')

# 5. Check the new correlations
new_correlations = multivariate_df[['y', 'freight_value', 'avg_price', 'active_sellers']].corr()['y']
print("--- New Correlation Results ---")
print(new_correlations)
```

```python
m_multi = Prophet()
m_multi.add_regressor('freight_value')
m_multi.fit(multivariate_df)
future_multi = m_multi.make_future_dataframe(periods=90)
future_multi['freight_value'] = multivariate_df['freight_value'].iloc[-1]
forecast_multi = m_multi.predict(future_multi)
```

```python
import itertools
param_grid = {  
    'changepoint_prior_scale': [0.001, 0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 10.0],
}
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
maes = []
for params in all_params:
    m = Prophet(**params).fit(prophet_df) 
    
    df_cv = cross_validation(m, initial='365 days', period='30 days', horizon='90 days', parallel="processes")
    
    df_p = performance_metrics(df_cv, rolling_window=1)
    maes.append(df_p['mae'].values[0])
    tuning_results = pd.DataFrame(all_params)
tuning_results['mae'] = maes
best_params = all_params[tuning_results['mae'].argmin()]

print("--- Best Hyperparameters Found ---")
print(best_params)
```

```python
m_prophet = Prophet(
    changepoint_prior_scale=0.5, 
    seasonality_prior_scale=0.01
)
m_prophet.add_country_holidays(country_name='BR')
m_prophet.fit(prophet_df)
forecast_train = m_prophet.predict(prophet_df)
```
